```python
import pandas as pd
```

```python
filename = "w241_matches_only.csv"
```

```python
df = pd.read_csv(filename)
```

```python
df.head()
```

```python
df_has_potential_match = df.copy()

df_has_potential_match_y = df_has_potential_match.copy()
df_has_potential_match_y["dup_cnts"] = df_has_potential_match_y["Matches"]
df_has_potential_match_y = df_has_potential_match_y.reindex(
    df_has_potential_match_y.index.repeat(
        df_has_potential_match_y["dup_cnts"]
    )
)
df_has_potential_match_y["match"] = 1

df_has_potential_match_n = df_has_potential_match.copy()
df_has_potential_match_n["dup_cnts"] = (
    df_has_potential_match_n["Swipes"] - df_has_potential_match_n["Matches"]
)
df_has_potential_match_n = df_has_potential_match_n.reindex(
    df_has_potential_match_n.index.repeat(
        df_has_potential_match_n["dup_cnts"]
    )
)
df_has_potential_match_n["match"] = 0
```

```python
df_expanded = pd.concat(
    [
        df_has_potential_match_y,
        df_has_potential_match_n,
        # df_no_match,
    ]
).reset_index(drop=True)
```

```python
df_expanded["day_name"] = pd.to_datetime(df_expanded["Date"] + "/23").dt.day_name()
df_expanded["weekday"] = pd.to_datetime(df_expanded["Date"] + "/23").dt.weekday
```

```python
(
    df_expanded
    .sort_values(["Owner", "Profile Name", "City", "Control vs Treatment", "Date"])
    .head(50)
)
```

```python
df_expanded.to_csv(filename.replace(".csv", "_expanded.csv"), index=False)
```
